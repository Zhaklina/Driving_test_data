```python
def scrape (date):
    
    from selenium import webdriver
    from selenium.webdriver.common.keys import Keys
    from bs4 import BeautifulSoup
    import re
    import pandas as pd
    from tabulate import tabulate
    import os
    import requests

    capabilities = webdriver.DesiredCapabilities().FIREFOX
    capabilities["marionette"] = True
    
    url = "http://78.157.22.16/aicrez/default.aspx?ctl00$MainContent$textDatumOd="+date
  

    fireFoxOptions = webdriver.FirefoxOptions()
    fireFoxOptions.set_headless()
    driver = webdriver.Firefox(firefox_options=fireFoxOptions)

    driver.implicitly_wait(3)
    driver.get(url)

    python_button = driver.find_element_by_id('ctl00_MainContent_LinkButton3') #FHSU
    python_button.click() 

    html=BeautifulSoup(driver.page_source, 'lxml')
    table_id = 'ctl00_MainContent_GridView1'
    rezultati = html.find('table', attrs={'id': table_id})

    df = pd.read_html(str(rezultati))
    df = df[0]
    df['Датум'] = (date)
    df.drop(['Реден бр.'], axis=1, inplace=True)
    df.rename({'Име, име на родител и презиме': 'Кандидат'}, axis=1, inplace=True)

    #df.to_csv(r'/Users/macbookpro/Desktop/scraped.csv') #site ke gi zachuvuva so isto ime, mu dodadov +date posle path-ot, ama taka zachuvuva prazni fajlovi :((

    print ('Done, here you are')
    return df
```
